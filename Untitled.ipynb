```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
interval = 50
folder = './Drive/record/'
colors = sns.color_palette("hls", 8)

f0 = '0.70_0.99_normal_steps'
df0 = pd.read_csv(folder+f0+'.csv')
df0.columns = [f0]
f2 = '0.70_0.99_n_ethical_steps'
df2 = pd.read_csv(folder+f2+'.csv')
df2.columns = [f2]
f3 = '0.70_0.99_p_ethical_steps'
df3 = pd.read_csv(folder+f3+'.csv')
df3.columns = [f3]

df = pd.concat([df0, df2, df3], axis=1)
mean = df.rolling(interval).mean()
mean.plot()

plt.plot()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
interval = 50
folder = './Drive/record/'
colors = sns.color_palette("hls", 8)

f0 = '2.00_0.20_normal_cat_hits'
df0 = pd.read_csv(folder+f0+'.csv')
df0.columns = [f0]
f1 = '2.00_0.20_human_n_cat_hits'
df1 = pd.read_csv(folder+f1+'.csv')
df1.columns = [f1]
f2 = '1.00_0.20_n_ethical_cat_hits'
df2 = pd.read_csv(folder+f2+'.csv')
df2.columns = [f2]
f3 = '2.00_0.20_human_p_cat_hits'
df3 = pd.read_csv(folder+f3+'.csv')
df3.columns = [f3]
f4 = '1.00_0.20_p_ethical_cat_hits'
df4 = pd.read_csv(folder+f4+'.csv')
df4.columns = [f4]

df = pd.concat([df0, df1, df2, df3, df4], axis=1)
mean = df.rolling(interval).mean()
mean.plot()
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)

```
